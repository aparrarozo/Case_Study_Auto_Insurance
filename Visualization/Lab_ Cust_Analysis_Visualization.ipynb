```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
```

```python
ca_plotting = pd.read_csv('Data_Marketing_Customer_Analysis_Round2.csv')
```

```python
ca_plotting.head()
```

```python
def drop_columns(ca_plotting) :
    ca_plotting.drop(columns=["Unnamed: 0"], inplace=True)
    return ca_plotting
```

```python
drop_columns(ca_plotting)
```

```python
ca_objects = ca_plotting.select_dtypes(include=['object'])
```

```python
ca_objects
```

```python
ca_floats = ca_plotting.select_dtypes(include=['float64'])
ca_floats
```

```python
ca_integers = ca_plotting.select_dtypes(include=['int64'])
```

```python
ca_integers
```

```python
ca_numerical = pd.concat([ca_integers,ca_floats], axis=1)
```

```python
ca_numerical
```

```python
plt.figure(figsize=(20,12))
for i, column in enumerate(ca_numerical.columns,1):
    plt.subplot(3,3,i)
    sns.histplot(ca_numerical[column])
```

```python
### Normalize data
normalized_numerical=(ca_numerical-ca_numerical.min())/(ca_numerical.max()-ca_numerical.min())
```

```python
plt.figure(figsize=(20,12))
for i, column in enumerate(normalized_numerical.columns,1):
    plt.subplot(3,3,i)
    sns.histplot(normalized_numerical[column])
```

```python
###Standardize data
z_score_numerical = (ca_numerical-ca_numerical.mean())/ca_numerical.std()
```

```python
plt.figure(figsize=(20,12))
for i, column in enumerate(z_score_numerical.columns,1):
    plt.subplot(3,3,i)
    sns.histplot(z_score_numerical[column])
```
